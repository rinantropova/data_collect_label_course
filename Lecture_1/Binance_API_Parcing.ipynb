```python
import requests
import json
import pandas as pd
import datetime as dt
```

```python
url = 'https://api.binance.com/api/v3/uiKlines'
symbol = 'BTCBUSD'
interval = '1h'
start = str(int(dt.datetime(2023, 1, 1).timestamp()*1000))
end = str(int(dt.datetime(2023, 12, 1).timestamp()*1000))
```

```python
params = {
    'symbol': symbol,
    'interval': interval,
    'startTime': start,
    'endTime': end
}
```

```python
response = requests.get(url, params=params).text
```

```python
response
```

```python
json_res = json.loads(response)
```

```python
json_res
```

```python
df = pd.DataFrame(json_res)
df.columns = ['datetime', 'open', 'high', 'low', 'close', 'volume', 'clode_time', 'qav', 'num_trades', 'taker_base_vol', 'taker_quote_vol', 'ignore']
```

```python
df.index = [dt.datetime.fromtimestamp(x/1000.0) for x in df.datetime]
```
